```python
import matplotlib.pyplot as plt

# Points for the line
x_points = [1, 1, 3, 3]
y_points = [0, 1, 1, 0]

# Plotting the line
plt.plot(x_points, y_points, color='blue')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vertical Line Passing Through (1,0) and (3,0)')

# Setting the x and y axis limits
plt.xlim(0, 4)
plt.ylim(0, 1)

# Displaying the plot
plt.grid(True)
plt.show()
```
