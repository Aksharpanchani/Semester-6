```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def harvestin_rate(x,h):
    r = 1; k = 1000;
    return r*x*(1- (x/k)) - h

def euler_integrate(x0, h, dt, tmax):
    x = x0
    t_list = np.arange(0,tmax,dt)
    x_list = []
    for t in t_list:
        x_list.append(x)
        x = x + dt*harvestin_rate(x,h)
    return t_list, x_list


x = np.arange(0,1000,1)
plt.plot(x,harvestin_rate(x,0),label='h=0')
plt.plot(x,harvestin_rate(x,100),label='h=100')
plt.plot(x,harvestin_rate(x,500),label='h=500')
plt.legend()
plt.xlabel('x')
plt.ylabel('$\dot{x}$')
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.show()

t_list, x_list = euler_integrate(10, 0, 0.0001, 0.5)
plt.plot(t_list,x_list,label='h=0')
t_list, x_list = euler_integrate(10, 100, 0.0001, 0.5)
plt.plot(t_list,x_list,label='h=100')
t_list, x_list = euler_integrate(10, 500, 0.0001, 0.5)
plt.plot(t_list,x_list,label='h=500')
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.legend()
plt.xlabel('t')
plt.ylabel('x')
plt.show()



def analytical_sol(t):
    c = 99
    return 1000.0*np.exp(t)/(c + np.exp(t))

# plot analytical solution and numerical solution in same plot for h=0 
t_list, x_list = euler_integrate(10, 0, 0.0001, 10)
plt.plot(t_list,x_list,label='numerical')
t_list = np.arange(0,10,0.0001)
plt.plot(t_list,analytical_sol(t_list),label='analytical')
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.legend()
plt.xlabel('t')
plt.ylabel('x')
plt.show()



# plot relative error for h=0 and plot it against t
t_list, x_list = euler_integrate(10, 0, 0.0001, 10)
plt.plot(t_list,(x_list-analytical_sol(t_list))/analytical_sol(t_list))
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.xlabel('t')
plt.ylabel('relative error')
plt.show()

print(analytical_sol(t_list)[0]) 
print(x_list[0])
print(analytical_sol(0))

    
```

```python
def agricultural_innovation_rate(X,a):
    return (X+a)*(1-X)

# plot agricultural-innovation-rate vs x for a = 0, 0.2 and 0.5
x = np.arange(0,1,0.01)
plt.plot(x,agricultural_innovation_rate(x,0),label='a=0')
plt.plot(x,agricultural_innovation_rate(x,0.2),label='a=0.2')
plt.plot(x,agricultural_innovation_rate(x,0.5),label='a=0.5')
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
# plt.yscale('log')
plt.legend()
plt.xlabel('X')
# y-label is X_dot (dot above X)
plt.ylabel('$\dot{X}$') 
plt.show()

def euler_integrate2(x0, a, dt, tmax):
    x = x0
    t_list = np.arange(0,tmax,dt)
    x_list = []
    for t in t_list:
        x_list.append(x)
        x = x + dt*agricultural_innovation_rate(x,a)
    return t_list, x_list
# plot x vs t for x(0) = 0.001 for a = 0, 0.2 and 0.5
t_list, x_list = euler_integrate2(0.001, 0, 0.00001, 20)
plt.plot(t_list,x_list,label='a=0')
t_list, x_list = euler_integrate2(0.001, 0.2, 0.00001, 20)
plt.plot(t_list,x_list,label='a=0.2')
t_list, x_list = euler_integrate2(0.001, 0.5, 0.00001, 20)
plt.plot(t_list,x_list,label='a=0.5')
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.xticks(np.arange(0, 21, 2))
plt.legend()
plt.xlabel('T')
plt.ylabel('X')
plt.show()

```
